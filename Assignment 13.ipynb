Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively

```python
#IndexErrob
try:
    numbers = [1, 2, 3]
    print(numbers[3])
except IndexError as e:
    print(e)
```

```python
#KeyError
try:
    fruits = {"apple": 50, "banana": 40, "orange": 60}
    print(fruits["mango"])
except KeyError as e:
    print("Mango is not in the dictionary")
```

Q5. Explain ImportError. What is ModuleNotFoundError?

An ImportError is an exception that is raised when Python cannot import a module or a name from a module.

A ModuleNotFoundError is a subclass of ImportError that is raised when the module being imported does not exist or cannot be found.

```python
#ModuleNotFoundError
try:
    import amar
except ModuleNotFoundError as e:
    print(e)
```
